```python
#Replacing Outliers
Q1=df['Tenure'].quantile(0.10)
Q3=df['Tenure'].quantile(0.90)
IQR=Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 - (whisker_width*IQR)
upper_whisker = Q3 + (whisker_width*IQR)
index=df['Tenure'][(df['Tenure']>upper_whisker)|(df['Tenure']<lower_whisker)].index
df.drop(index,inplace=True)
```
